```r
library(reshape2)
library(ggplot2)
library(factoextra)
library(tidyverse)
```

```r
data_path <- "./data/data_clean_back.csv"
data <- read.csv(data_path)
pca <- prcomp(data, center = TRUE, scale. = TRUE)

# # 获取主成分
# pca$rotation

# # 获取方差贡献率
# pca$sdev^2 / sum(pca$sdev^2)

# # 获取主成分得分
# pca$x
# print(str(pca))
# 导入 ggplot2 包
library(ggplot2)

# 创建数据帧，其中包含主成分得分和原始数据
pca_data <- data.frame(pca$x, data)

# 创建散点图，其中每个点代表一个原始数据点，颜色表示主成分 1 的得分
ggplot(pca_data, aes(x = PC1, y = PC2, color = PC1)) +
  geom_point()
library(factoextra)

# 创建主成分得分图
fviz_pca_var(pca, col.var = "black", 
             gradient.cols = c("blue", "red"))
# 创建主成分载荷图
fviz_pca_var(pca, col.var = "black", 
             gradient.cols = c("blue", "red"),
             legend.title = "Variable")
# 创建累积方差图
fviz_eig(pca)
# 创建带有置信椭圆的主成分得分图
fviz_pca_var(pca, col.var = "black", 
             gradient.cols = c("blue", "red"),
             addEllipses = TRUE, ellipse.type = "confidence")
# 根据分组上色并绘制95%置信区间
fviz_pca_ind(pca, col.ind=data$conditions, mean.point=F, addEllipses = T, legend.title="Groups", ellipse.type="confidence", ellipse.level=0.95)
```

```r
data_path <- "./data/data_clean.csv"
data <- read.csv(data_path)
data <- data %>%
  select_if(function(x) var(x) != 0)
data_t <- t(data)
variableL <- ncol(data_t)
data_t
if(data_path != "") {
  sample <- read.csv(data_path,header = T, row.names=1,sep="\t")
  sample <- sample %>%
    select_if(function(x) var(x) != 0)
  data_t_m <- merge(data_t, sample, by=0)
  rownames(data_t_m) <- data_t_m$Row.names
  data_t <- data_t_m[,-1]
}
# By default, prcomp will centralized the data using mean.
# Normalize data for PCA by dividing each data by column standard deviation.
# Often, we would normalize data.
# Only when we care about the real number changes other than the trends,
# `scale` can be set to TRUE.
# We will show the differences of scaling and un-scaling effects.
pca <- prcomp(data_t[,1:variableL], scale=T)

# sdev: standard deviation of the principle components.
# Square to get variance
# percentVar <- pca$sdev^2 / sum( pca$sdev^2)

# To check what's in pca
print(str(pca))
```

```r
fviz_eig(pca, addlabels = TRUE)
```

```r
fviz_pca_ind(pca, repel=T) 
```

```r
fviz_pca_ind(pca, col.ind=data_t$conditions, mean.point=F, addEllipses = T, legend.title="Groups")
```
