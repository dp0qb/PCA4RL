```r
library(reshape2)
library(ggplot2)
library(factoextra)
library(tidyverse)
library("corrplot")
library("FactoMineR");
library("factoextra");

remove_0_var_col <- function(data) {
  data <- data %>%
  select_if(function(x) var(x) != 0)
  data
}
```

```r
data_path <- "./data/data_clean_back.csv"
data <- read.csv(data_path)

data
# pca <- prcomp(data, center = TRUE, scale. = TRUE)
# PCA计算
res.pca <- PCA(data[, -1], graph = FALSE)
# 提取变量的分析结果
var <- get_pca_var(res.pca)
# var
# Coordinates of variables
# head(var$coord, 12)
# col.var设定线条颜色
fviz_pca_var(res.pca, col.var = "black")
# head(var$cos2)

# is.corr表示输入的矩阵不是相关系数矩阵
corrplot(var$cos2, is.corr=FALSE)

# Total cos2 of variables on Dim.1 and Dim.2
# 在第一第二主成分是显示结果（通过值的叠加显示）
fviz_cos2(res.pca, choice = "var", axes = 1:2)

# Color by cos2 values: quality on the factor map
fviz_pca_var(res.pca, col.var = "cos2",
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"), 
             repel = TRUE # Avoid text overlapping
             )
# ggsave(p, "my_plot.png")

# Quality of individuals
# head(ind$cos2)
p <- fviz_pca_ind(res.pca,
             # show points only (nbut not "text") 只显示点而不显示文本，默认都显示
             geom.ind = "point", 
             # 设定分类种类
             col.ind = data$Deposit.type,
             # 设定颜色
             palette = rainbow(8),
             # 添加椭圆 Concentration ellipses
             addEllipses = TRUE,
             legend.title = "Groups",
             )

```
